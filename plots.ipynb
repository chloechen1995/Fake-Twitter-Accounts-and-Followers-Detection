```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('final_df_sample_4.csv', index_col = 0)
```

```python
name = ['followers_count', 'following', 'statuses_count', 'ff_ratio', 'age','following_rate', 'tweet_rate',
         'bidirectional_links', 'pagerank', 'reputation', 'url_unique_ratio','hashtag_ratio', 'username_unique_ratio',
        'similarity_ratio_1', 'similarity_ratio_2', 'tweet_time_mean', 'label_model']
```

```python
new_data = data[name]
```

```python
def reject_outliers(data, m = 2.):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else 0.
    return data[s<m]
```

```python
#excl = reject_outliers(form.ix[:,0])
#form[form['followers_count'].isin(excl)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (12, 28*4), dpi = 100)
for i in range(16):
    check = new_data.ix[:, i] != -1
    form = pd.DataFrame(new_data.ix[:, i][check])
    form['label_model'] = new_data['label_model'][check]
    excl = reject_outliers(form.ix[:,0])
    form = form[form.ix[:,0].isin(excl)]
    form.boxplot(by = 'label_model')
    # print str(i) + "succeed"
```
